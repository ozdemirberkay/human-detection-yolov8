```python
import matplotlib.pyplot as plt
from PIL import Image
import matplotlib.patches as patches
```

```python
def get_image_path_from_name(file_name):
    return f"./datasets/images/val/{file_name}.jpg"

def get_label_path_from_name(file_name):
    return f"./datasets/labels/val/{file_name}.txt"
```

```python
def draw_bounding_boxes(file_name):

    # Get image
    image = Image.open(get_image_path_from_name(file_name))
    img_width, img_height = image.size

    # Open label file
    with open(get_label_path_from_name(file_name), "r") as label_file:

        # Create a new figure
        fig, ax = plt.subplots()
        ax.imshow(image)

        for line in label_file:
            parts = line.strip().split()

            class_id, x_center, y_center, width, height = map(float, parts)

            # Get pixels from coordinates
            xmin = (x_center - width / 2) * img_width
            ymin = (y_center - height / 2) * img_height
            xmax = (x_center + width / 2) * img_width
            ymax = (y_center + height / 2) * img_height

            # Draw rectangle
            rect = patches.Rectangle((xmin, ymin), xmax - xmin, ymax - ymin, linewidth=2, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

        # Show image
        plt.show()
```

```python
draw_bounding_boxes("0a88602cbfb40e0f")
```
